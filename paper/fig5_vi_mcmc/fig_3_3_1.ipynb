```python
import svgutils.transform as sg
import sys

#create new SVG figure
fig = sg.SVGFigure(width="14cm", height="3cm")

# load matpotlib-generated figures
fig1 = sg.fromfile('figs/default_sampling_times.svg')
fig2 = sg.fromfile('figs/two_moons_failure.svg')


# fig12_legend = sg.fromfile('method_legend_reduced.svg')

# # get the plot objects
plot1 = fig1.getroot()
plot2 = fig2.getroot()
plot1.moveto(0, 0)
plot2.moveto(190, 3)


# # add text labels
txt1 = sg.TextElement(5,15, "a", size=12, weight="bold", font="Arial")
txt2 = sg.TextElement(180,15, "b", size=12, weight="bold", font="Arial")

# # append plots and labels to figure
fig.append([plot1, plot2])
fig.append([txt1, txt2])

fig.save("figs/fig3_3_1.svg")

from IPython.display import SVG

SVG('figs/fig3_3_1.svg')
```
